```python
sample_file_path = os.path.join(audio_file_path, "1-137-A-32.wav")
```

```python
y, sr = librosa.load(sample_file_path, sr=None)
print("y 音频波形: ",y)
print("sr 采样率 : ",sr)
```

```python
time = librosa.times_like(y, sr=sr)

plt.figure(figsize=(14, 4))
librosa.display.waveshow(y, sr=sr, alpha=0.6)
plt.title('Waveform of "1-137-A-32.wav"')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()
```
